```python
sns.countplot(df['rating']).unique()
```

```python
pd.set_option('display.max_rows', 1000)
```

```python
df_month = df['rating'].groupby([df.written_date.dt.year, df.written_date.dt.month]).agg('count')
df_month
#This is showing how many reviews came in over the years for each month. 
```

```python
df.groupby(pd.to_datetime(df.written_date).dt.month).count()
```

```python
df.groupby(pd.to_datetime(df['written_date']).dt.strftime('%b-%Y'), sort=False)['rating'].size().plot()

#just a quick look at when the most reviews came in#
```
